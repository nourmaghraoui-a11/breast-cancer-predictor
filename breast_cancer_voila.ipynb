```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML
import os
os.environ['LOKY_MAX_CPU_COUNT'] = '10'

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score, classification_report

df = pd.read_csv('breast-cancer.csv')
y  = df['diagnosis'].map({'B': 0, 'M': 1}).astype(int).values
X  = df.drop(columns=['id', 'diagnosis']).values
feature_names = df.drop(columns=['id', 'diagnosis']).columns.tolist()

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=21, stratify=y
)
pipeline = Pipeline([('scaler', StandardScaler()), ('knn', KNeighborsClassifier(n_neighbors=2))])
pipeline.fit(X_train, y_train)
y_pred   = pipeline.predict(X_test)
acc      = accuracy_score(y_test, y_pred)
rep      = classification_report(y_test, y_pred, target_names=['Benign','Malignant'], output_dict=True)
defaults = df.drop(columns=['id', 'diagnosis']).median().to_dict()

# ══════════════════════════
# CSS
# ══════════════════════════
display(HTML("""
<style>
@import url('https://fonts.googleapis.com/css2?family=Space+Mono:wght@400;700&display=swap');

body, .jp-Notebook, .jp-Cell, .jp-OutputArea-output, #jp-main-content {
  background: #0f1117 !important; margin:0 !important; padding:0 !important;
}
.jp-Cell-inputWrapper { display: none !important; }
.jp-Cell { padding: 0 !important; border: none !important; }
* { box-sizing: border-box; }

.rx-header {
  background:#13151f; border-bottom:1px solid #1e2535;
  padding:16px 28px; display:flex; align-items:center; gap:14px;
  font-family:'Space Mono',monospace;
}
.rx-badge {
  width:44px; height:44px; border-radius:50%;
  background:linear-gradient(135deg,#6366f1,#f43f5e);
  display:flex; align-items:center; justify-content:center; font-size:22px; flex-shrink:0;
}
.rx-title { color:#f1f5f9; font-size:18px; font-weight:700; letter-spacing:1px; }
.rx-sub   { color:#64748b; font-size:10px; letter-spacing:1.5px; margin-top:3px; }

.rx-metrics { display:flex; gap:12px; padding:16px 28px; background:#0f1117; }
.rx-metric {
  flex:1; background:#13151f; border:1px solid #1e2535;
  border-radius:10px; padding:14px; text-align:center; font-family:'Space Mono',monospace;
}
.rx-metric-val { color:#6366f1; font-size:20px; font-weight:700; }
.rx-metric-lbl { color:#475569; font-size:9px; letter-spacing:1.5px; margin-top:4px; }

/* Tab widget styling */
.jupyter-widgets.widget-tab > .p-TabBar .p-TabBar-tab,
.jupyter-widgets.widget-tab > .lm-TabBar .lm-TabBar-tab {
  font-family:'Space Mono',monospace !important;
  font-size:11px !important;
  background:transparent !important;
  color:#64748b !important;
  border:1px solid #1e2535 !important;
  border-radius:8px 8px 0 0 !important;
  padding:8px 16px !important;
  margin-right:4px !important;
}
.jupyter-widgets.widget-tab > .p-TabBar .p-TabBar-tab.p-mod-current,
.jupyter-widgets.widget-tab > .lm-TabBar .lm-TabBar-tab.lm-mod-current {
  background:#6366f115 !important;
  color:#6366f1 !important;
  border-color:#6366f155 !important;
  font-weight:700 !important;
}
.jupyter-widgets.widget-tab > .widget-tab-contents {
  background:#0f1117 !important;
  border:1px solid #1e2535 !important;
  padding:14px !important;
}

/* Feature card */
.rx-card {
  background:#13151f; border:1px solid #1e2535;
  border-radius:10px; padding:12px 14px;
}
.rx-card-top {
  display:flex; justify-content:space-between;
  align-items:center; margin-bottom:6px;
}
.rx-card-name { font-family:'Space Mono',monospace; font-size:11px; color:#94a3b8; }
.rx-bar-bg    { height:4px; background:#1e2535; border-radius:4px; overflow:hidden; margin-bottom:4px; }

/* Input inside card */
.widget-bounded-float-text input,
.widget-float-text input {
  background:#0f1117 !important;
  border:1px solid #2d3748 !important;
  border-radius:6px !important;
  color:#6366f1 !important;
  font-family:'Space Mono',monospace !important;
  font-size:12px !important;
  font-weight:700 !important;
  padding:4px 8px !important;
  text-align:right !important;
  width:100px !important;
}
.widget-bounded-float-text input:focus,
.widget-float-text input:focus {
  border-color:#6366f1 !important;
  outline:none !important;
  box-shadow:0 0 0 2px #6366f122 !important;
}

/* Result panel */
.rx-result-empty {
  background:#13151f; border:1px dashed #1e2535;
  border-radius:12px; padding:32px 16px; text-align:center;
  font-family:'Space Mono',monospace; color:#334155; font-size:12px;
}
.rx-result-b { background:#22c55e12; border:1px solid #22c55e44; border-radius:12px; padding:18px; text-align:center; margin-bottom:10px; }
.rx-result-m { background:#f43f5e12; border:1px solid #f43f5e44; border-radius:12px; padding:18px; text-align:center; margin-bottom:10px; }
.rx-diag-lbl { font-family:'Space Mono',monospace; font-size:10px; color:#64748b; letter-spacing:2px; margin-bottom:6px; }
.rx-diag-b   { color:#22c55e; font-family:'Space Mono',monospace; font-size:22px; font-weight:700; }
.rx-diag-m   { color:#f43f5e; font-family:'Space Mono',monospace; font-size:22px; font-weight:700; }
.rx-info-box { background:#13151f; border:1px solid #1e2535; border-radius:10px; padding:12px; margin-bottom:10px; }
.rx-info-ttl { font-family:'Space Mono',monospace; font-size:9px; color:#64748b; letter-spacing:2px; margin-bottom:8px; }
.rx-proba-grid { display:grid; grid-template-columns:1fr 1fr; gap:8px; }
.rx-pb { background:#13151f; border:1px solid #22c55e33; border-radius:8px; padding:8px; text-align:center; }
.rx-pm { background:#13151f; border:1px solid #f43f5e33; border-radius:8px; padding:8px; text-align:center; }
.rx-footer {
  display:flex; gap:20px; padding:10px 28px;
  border-top:1px solid #1e2535; background:#13151f;
  flex-wrap:wrap; align-items:center; margin-top:8px;
}
.rx-fd { display:flex; align-items:center; gap:6px; font-family:'Space Mono',monospace; font-size:10px; color:#475569; }
.dot   { width:8px; height:8px; border-radius:50%; }
</style>
"""))

# ── Header ──
display(HTML(f"""
<div class='rx-header'>
  <div class='rx-badge'>🔬</div>
  <div>
    <div class='rx-title'>BREAST CANCER PREDICTOR</div>
    <div class='rx-sub'>KNN · k=2 · StandardScaler · Wisconsin Dataset · Accuracy {acc:.1%}</div>
  </div>
</div>
"""))

# ── Metrics ──
display(HTML(f"""
<div class='rx-metrics'>
  <div class='rx-metric'><div class='rx-metric-val'>{acc:.1%}</div><div class='rx-metric-lbl'>ACCURACY</div></div>
  <div class='rx-metric'><div class='rx-metric-val'>{rep['Malignant']['f1-score']:.3f}</div><div class='rx-metric-lbl'>F1 MALIGNANT</div></div>
  <div class='rx-metric'><div class='rx-metric-val'>{rep['Benign']['recall']:.3f}</div><div class='rx-metric-lbl'>RECALL BENIGN</div></div>
  <div class='rx-metric'><div class='rx-metric-val'>{X_train.shape[0]}</div><div class='rx-metric-lbl'>TRAIN SAMPLES</div></div>
</div>
"""))

# ══════════════════════════
# Feature inputs inside cards
# ══════════════════════════
feat_inputs = {}
for feat in feature_names:
    col = df[feat]
    feat_inputs[feat] = widgets.BoundedFloatText(
        value=round(defaults[feat], 4),
        min=round(col.min(), 6),
        max=round(col.max(), 6),
        step=round((col.max()-col.min())/200, 6),
        description='',
        layout=widgets.Layout(width='110px', height='32px')
    )

def get_label(feat):
    return feat.replace('_mean','').replace('_se','').replace('_worst','').replace('_',' ').title()

def make_card(feat, bar_color):
    col  = df[feat]
    val  = feat_inputs[feat].value
    pct  = max(0, min(100, (val - col.min()) / (col.max() - col.min()) * 100))
    name_html = widgets.HTML(
        f"<div class='rx-card-name'>{get_label(feat)}</div>"
    )
    bar_html = widgets.HTML(
        f"<div class='rx-bar-bg'>"
        f"<div style='height:100%;width:{pct:.1f}%;background:{bar_color};border-radius:4px'></div>"
        f"</div>"
    )
    top_row = widgets.HBox(
        [name_html, feat_inputs[feat]],
        layout=widgets.Layout(justify_content='space-between', align_items='center', width='100%')
    )
    card = widgets.VBox(
        [top_row, bar_html],
        layout=widgets.Layout(
            border='1px solid #1e2535', border_radius='10px',
            padding='12px 14px', background='#13151f', width='100%'
        )
    )
    return card

def make_group_tab(feats, bar_color):
    left_cards  = [make_card(f, bar_color) for f in feats[0::2]]
    right_cards = [make_card(f, bar_color) for f in feats[1::2]]
    left_col  = widgets.VBox(left_cards,  layout=widgets.Layout(flex='1', gap='10px', min_width='0'))
    right_col = widgets.VBox(right_cards, layout=widgets.Layout(flex='1', gap='10px', min_width='0'))
    return widgets.HBox(
        [left_col, right_col],
        layout=widgets.Layout(gap='10px', width='100%')
    )

mean_feats  = [f for f in feature_names if '_mean'  in f]
se_feats    = [f for f in feature_names if '_se'    in f]
worst_feats = [f for f in feature_names if '_worst' in f]

feat_tabs = widgets.Tab(children=[
    make_group_tab(mean_feats,  '#6366f1'),
    make_group_tab(se_feats,    '#0ea5e9'),
    make_group_tab(worst_feats, '#f43f5e'),
])
feat_tabs.set_title(0, 'Mean Features')
feat_tabs.set_title(1, 'Standard Error (SE)')
feat_tabs.set_title(2, 'Worst Features')
feat_tabs.layout = widgets.Layout(flex='1', min_width='0')

# ══════════════════════════
# Result panel
# ══════════════════════════
result_out = widgets.Output()
with result_out:
    display(HTML("<div class='rx-result-empty'>🧬<br><br>Set features and click<br>Run Prediction</div>"))

btn_predict = widgets.Button(
    description='▶  RUN PREDICTION',
    layout=widgets.Layout(flex='1', height='44px'),
    style={'button_color':'#6366f1', 'font_weight':'bold'}
)
btn_reset = widgets.Button(
    description='↺  RESET',
    layout=widgets.Layout(width='100px', height='44px'),
    style={'button_color':'#13151f'}
)

def on_predict(b):
    sample = np.array([[feat_inputs[f].value for f in feature_names]])
    pred   = pipeline.predict(sample)[0]
    proba  = pipeline.predict_proba(sample)[0]
    conf   = int(max(proba) * 100)
    scaled = pipeline.named_steps['scaler'].transform(sample)[0]
    top3   = np.abs(scaled).argsort()[-3:][::-1]
    bar_c  = '#f43f5e' if pred==1 else '#22c55e'

    rows = "".join([
        f"<tr>"
        f"<td style='font-family:Space Mono,monospace;color:#94a3b8;font-size:11px;padding:4px 6px'>{get_label(feature_names[i])}</td>"
        f"<td style='font-family:Space Mono,monospace;color:#6366f1;font-size:11px;text-align:right;padding:4px 6px'>{feat_inputs[feature_names[i]].value:.4f}</td>"
        f"<td style='font-family:Space Mono,monospace;color:#475569;font-size:10px;padding:4px 6px'>z={scaled[i]:+.2f}</td>"
        f"</tr>"
        for i in top3
    ])

    if pred == 1:
        card_cls, val_cls, icon, label = 'rx-result-m', 'rx-diag-m', '⚠', 'MALIGNANT'
    else:
        card_cls, val_cls, icon, label = 'rx-result-b', 'rx-diag-b', '✓', 'BENIGN'

    result_out.clear_output(wait=True)
    with result_out:
        display(HTML(f"""
        <div class='{card_cls}'>
          <div class='rx-diag-lbl'>DIAGNOSIS</div>
          <div class='{val_cls}'>{icon} {label}</div>
          <div style='font-family:Space Mono,monospace;font-size:10px;color:#64748b;margin:10px 0 5px'>CONFIDENCE: {conf}%</div>
          <div style='height:8px;background:#1e2535;border-radius:4px;overflow:hidden'>
            <div style='height:100%;width:{conf}%;background:{bar_c};border-radius:4px'></div>
          </div>
        </div>
        <div class='rx-info-box'>
          <div class='rx-info-ttl'>TOP 3 INFLUENTIAL FEATURES</div>
          <table style='width:100%;border-collapse:collapse'>{rows}</table>
        </div>
        <div class='rx-proba-grid'>
          <div class='rx-pb'>
            <div style='color:#22c55e;font-family:Space Mono,monospace;font-size:14px;font-weight:700'>Benign</div>
            <div style='color:#475569;font-family:Space Mono,monospace;font-size:10px'>{proba[0]:.1%}</div>
          </div>
          <div class='rx-pm'>
            <div style='color:#f43f5e;font-family:Space Mono,monospace;font-size:14px;font-weight:700'>Malignant</div>
            <div style='color:#475569;font-family:Space Mono,monospace;font-size:10px'>{proba[1]:.1%}</div>
          </div>
        </div>
        """))

def on_reset(b):
    for feat in feature_names:
        feat_inputs[feat].value = round(defaults[feat], 4)
    result_out.clear_output(wait=True)
    with result_out:
        display(HTML("<div class='rx-result-empty'>🧬<br><br>Features reset to median</div>"))

btn_predict.on_click(on_predict)
btn_reset.on_click(on_reset)

right_panel = widgets.VBox(
    [
        widgets.HTML("<div style='font-family:Space Mono,monospace;font-size:10px;color:#64748b;letter-spacing:2px;margin-bottom:12px'>PREDICTION RESULT</div>"),
        result_out,
        widgets.HBox([btn_predict, btn_reset], layout=widgets.Layout(margin='12px 0 0 0', gap='8px'))
    ],
    layout=widgets.Layout(width='310px', min_width='280px', padding='0 0 0 20px')
)

display(widgets.HBox(
    [feat_tabs, right_panel],
    layout=widgets.Layout(
        padding='0 28px 20px', align_items='flex-start', background='#0f1117'
    )
))

display(HTML("""
<div class='rx-footer'>
  <div class='rx-fd'><div class='dot' style='background:#6366f1'></div>Mean Features: 10</div>
  <div class='rx-fd'><div class='dot' style='background:#0ea5e9'></div>Standard Error: 10</div>
  <div class='rx-fd'><div class='dot' style='background:#f43f5e'></div>Worst Features: 10</div>
  <div style='margin-left:auto;font-family:Space Mono,monospace;font-size:10px;color:#334155'>⚠ For research purposes only — not a clinical tool</div>
</div>
"""))
```
